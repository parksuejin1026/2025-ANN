```python
import numpy as np

x = np.array([2,4,6,8,10,])
y = np.array([81, 93, 91 ,97, 120])

mx = np.mean(x)
my = np.mean(y)

print("x의 평균값:", mx)
print("y의 평균값:", my)

divisor = sum([(i - mx)**2 for i in x])

def top(x,mx,y,my) :
  d = 0
  for i in range(len(x)):
    d += (x[i] - mx) * (y[i] - my)
  return d

dividend = top(x, mx, y, my)

print("분모:", divisor)
print("분자:", dividend)

# 기울기 a를 구하는 공식입니다.
a = dividend / divisor

b = my - (mx*a)

print("기울기 a = ", a)
print("y절편 b = ", b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지




# x와 y의 평균값 계산
mx = np.mean(x)
my = np.mean(y)

print(f"x의 평균값: {mx}")
print(f"y의 평균값: {my}")

# 분모(divisor)와 분자(dividend) 계산
divisor = sum([(i - mx)**2 for i in x])

def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d

dividend = top(x, mx, y, my)

print(f"분모: {divisor}")
print(f"분자: {dividend}")

# 기울기(a)와 y절편(b) 계산
a = dividend / divisor
b = my - (mx * a)

print(f"기울기 a = {a}")
print(f"y절편 b = {b}")

# 계산된 기울기와 y절편으로 예측된 y값 계산
y_pred = a * x + b

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 원본 데이터는 점으로 표시
plt.scatter(x, y, color='blue', label='실제 요금 데이터')

# 예측된 선은 직선으로 표시
plt.plot(x, y_pred, color='red', label=f'예측선 (y = {a:.2f}x + {b:.2f})')

# 그래프 제목과 축 레이블 설정
plt.title('이동거리에 따른 택시 요금 예측', fontsize=16)
plt.xlabel('이동거리 (km)', fontsize=12)
plt.ylabel('금액 (원)', fontsize=12)

# 범례 표시
plt.legend()

# 그래프 격자 표시
plt.grid(True)

# 레이블이 잘리지 않도록 조정
plt.tight_layout()

# 그래프를 파일로 저장
plt.savefig('taxi_fare_regression_30_points.png')
```
