## 평균제곱오차

```
import numpy as np
```

```
# 가상의 기울기 A와 y절편 b를 정합니다.
fake_a = 2.3
fake_b = 79

# 이동거리 x와 금액 y의 넘파이 배열을 만듭니다.
x = np.array([2, 4, 6, 8, 5]) # 이상치
y = np.array([81, 93, 91, 97, 10])
```

```
# y = ax + b에 가상의 a,b 값을 대입한 결과를 출력하는 함수입니다.
def predict(x):
  return fake_a* x + fake_b

# 예측 값이 들어갈 빈 리스트를 만듭니다.
predict_result = []

# 모든 x값을 한 번씩 대입하여 predict_result 리스트를 완성합니다.
for i in range(len(x)):
   predict_result.append(predict(x[i]))
   print("이동거리 = %.f km, 실제점수 = %.f , 예측점수 = %.f" % (x[i] , y[i], predict(x[i])))
```

```
# 평균 제곱 오차 함수를 각 y 값에 대입하여 최종 값을 구하는 함수입니다.
n = len(x)
def mse(y, y_pred):
  return (1/n) * sum((y - y_pred) ** 2)

# 평균 제곱 오차 값을 출력합니다.
print("평균 제곱 오차 : " + str(mse(y,predict_result)))
```

```
import numpy as np

# 가상의 기울기 A와 y절편 b는 더 이상 필요하지 않습니다.
# 이동거리 x와 금액 y의 넘파이 배열을 만듭니다.
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 넘파이의 polyfit 함수를 사용해 최적의 직선(1차 다항식)을 찾습니다.
# polyfit(x, y, 1)에서 1은 1차 함수(직선)를 의미합니다.
a, b = np.polyfit(x, y, 1)

print(f"최종 기울기: {a:.4f}, 최종 y절편: {b:.4f}")
```

## 경사 하강법

```
import numpy as np
import matplotlib.pyplot as plt
```

```
# 이동 거리 x와 금액 y의 넘파이 배열을 만듭니다.
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```
# 기울기 a와 절편 b의 값을 초기화한다.
a = 0
b = 0

# 학습률을 정한다 learning rate
lr = 0.01

# 몇 번을 반복할 지 설정한다.
epochs = 4500

# x의 값이 총 몇 개인지 셉니다.
n = len(x)

# 경사 하강법을 시작합니다.
for i in range(epochs):

  y_pred = a * x + b
  error = y - y_pred

  a_diff = (2/n) * sum(-x * (error))
  b_diff = (2/n) * sum(-(error))

  a = a - lr * a_diff
  b = b - lr * b_diff

  if i % 50 == 0:
    print("epoch = %.f, 기울기 = %.04f, 절편 = %.04f" % (i, a, b))
```

```
# 앞서 구한 최종 a 값을 기울기 , b 값을 y 절편에 대입하여 그래프를 그린다.
y_pred = a * x + b

# 그래프 출력
plt.scatter(x,y)
plt.title("Gradient Descent") # 경사 하강법 이라는 뜻
plt.plot(x, y_pred, 'r')
plt.show()
```

## 텐서플로우 활용

```
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
```

```
model = Sequential()

# 출력 값, 입력 변수, 분석 방법에 맞게끔 모델을 설정합니다.
model.add(Dense(1, input_dim = 1, activation='linear'))

# 오차 수정을 위해 경사 하강법(sgd)을, 오차의 정도를 판단하기 위해 평균 제곱 오차(mse)를 사용합니다.
model.compile(optimizer='sgd', loss='mse')

# 오차를 최소화하는 과정을 2000번 반복한다.
model.fit(x, y, epochs = 2000)
```

```
plt.scatter(x,y)
plt.plot(x, model.predict(x), 'r') # 예측 결과를 그래프로 나타낸다.
plt.show()
```

## 다중 선형 회귀

```
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 api에서 필요한 함수들을 불러옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 숫자가 너무 크면 발산이 됨
x = np.array([[2, 5], [4,8], [6,10], [8,21]])
y = np.array([81, 93, 91, 97])

model = Sequential()

# 입력 변수가 2개(이동거리, 시간) 이므로 input_dim에 2를 입력한다.
model.add(Dense(1, input_dim = 2, activation = 'linear'))
model.compile(optimizer='sgd', loss = 'mse')

model.fit(x, y, epochs = 100)
```

```
import tensorflow as tf

distance = 7
minutes = 5

input_data = tf.constant([[distance, minutes]])
prediction = model.predict(input_data)[0][0]

print("%.f을 이동하고 %.f분을 움직이면 예상 금액은 %.02f원 입니다." %(distance, minutes, prediction))
```
